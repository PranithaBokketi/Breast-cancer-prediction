```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load the dataset
```

```python
# Load dataset
breast_cancer_data = load_breast_cancer()

# Convert to DataFrame
df = pd.DataFrame(breast_cancer_data.data, columns=breast_cancer_data.feature_names)
```

```python
# Add target column
df['label'] = breast_cancer_data.target
```

```python
# Display first and last rows
print(df.head())
```

```python
print(df.tail())
```

```python
# Exploratory Data Analysis (EDA)
```

```python
# Shape of data
print("Shape:", df.shape)
```

```python
# Info
print(df.info())
```

```python
# Missing values
print("Missing values:\n", df.isnull().sum())
```

```python
# Summary statistics
print(df.describe())
```

```python
# Target variable distribution
print(df['label'].value_counts())
```

```python
# Data Preprocessing
```

```python
X = df.drop(columns='label', axis=1)
Y = df['label']
```

```python
# Train-Test Split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
# Model Training
```

```python
model = LogisticRegression(max_iter=10000)  # Added max_iter to avoid convergence warning
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
```

```python
# Training accuracy
X_train_prediction = model.predict(X_train)
training_accuracy = accuracy_score(Y_train, X_train_prediction)
print("Accuracy on training data:", training_accuracy)

# Testing accuracy
X_test_prediction = model.predict(X_test)
testing_accuracy = accuracy_score(Y_test, X_test_prediction)
print("Accuracy on test data:", testing_accuracy)
```
